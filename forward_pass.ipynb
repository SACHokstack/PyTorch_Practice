
# LINEAR REGRESSION
## With y =2x+1

```python

N =10
D_in =1
D_out =1

import torch
X = torch.randn(N, D_in)

true_W = torch.tensor([[2.0]])
true_b = torch.tensor([1.0])
y_true =X @ true_W + true_b +torch.randn(N, D_out) *0.1
```

```python
W =torch.randn(D_in, D_out, requires_grad=True)
b =torch.randn(D_out, requires_grad=True)

print("Initial W:", W)
print("Initial b:", b)
```

```python
y_hat =X @ W + b

```

# Backward_Pass

```python
error =y_hat - y_true
squared_error =error **2
loss =squared_error.mean()
print("Initial loss:", loss.item())
```

```python
loss.backward()
```

```python
print(f"Gradient of W: {W.grad}")
print(f"Gradient of b: {b.grad}")
```

# theta t+1 = theta - n.deltaL

```python
learning_rate =0.01
w_new = W - learning_rate * W.grad
b_new = b - learning_rate * b.grad
```

```python
epochs =100

W,b = torch.randn(1,1, requires_grad=True), torch.randn(1, requires_grad=True)

for epoch in range(epochs):
    y_hat =X @ W + b
    loss =((y_hat - y_true) **2).mean()
    loss.backward()
    
    with torch.no_grad():
        W -= learning_rate * W.grad
        b -= learning_rate * b.grad
        
        W.grad.zero_()
        b.grad.zero_()

    if epoch %10 ==0:
        print(f"Epoch {epoch}, Loss: {loss.item()} , W: {W.item()}, b: {b.item()}")
print("Trained W:", W)
print("Trained b:", b)
```

# Using TORCH.NN

```python
import torch.nn as nn

class LinearRegressionModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.linear_layer =nn.Linear(input_dim, output_dim)
    
    def forward(self, x):
        return self.linear_layer(x)
model =LinearRegressionModel(input_dim =1, output_dim =1)
```

```python
print("Model Architecture:\n", model)
```

```python
import torch.optim as optim

learning_rate =0.01

optimizer =optim.Adam(model.parameters(), lr=learning_rate)

loss_fn =torch.nn.MSELoss()
```

```python
epoch =100

for epoch in range(epoch):
    y_hat =model(X)

    loss =loss_fn(y_hat, y_true)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch %10 ==0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")
        
```
